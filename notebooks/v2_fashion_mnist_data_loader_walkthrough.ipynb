## Step 1: Imports

Type the imports needed for loading Fashion-MNIST. The important source is `torchvision.datasets.FashionMNIST`, usually accessed as `datasets.FashionMNIST`.

```python
# TODO: type imports here
# from pathlib import Path
# from torch.utils.data import DataLoader
# from torchvision import datasets, transforms
from pathlib import Path

from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

## Step 2: Transform

Fashion-MNIST images are grayscale 28x28 images. `transforms.ToTensor()` converts one image into a tensor shaped `[1, 28, 28]`.

```python
# TODO: create the transform
# transform = transforms.Compose([transforms.ToTensor()])
# transform

transform = transforms.Compose([transforms.ToTensor()])
transform
```

## Step 3: Load Fashion-MNIST Datasets

This is the key difference from MNIST V1: use `datasets.FashionMNIST(...)` instead of `datasets.MNIST(...)`.

```python
# TODO: create train_dataset and test_dataset
# data_dir = Path("../data")
# train_dataset = datasets.FashionMNIST(root=str(data_dir), train=True, download=False, transform=transform)
# test_dataset = datasets.FashionMNIST(root=str(data_dir), train=False, download=False, transform=transform)
# len(train_dataset), len(test_dataset)

data_dir = Path("../data")

train_dataset = datasets.FashionMNIST(
    root=str(data_dir), train=True, download=False, transform=transform,
)

test_dataset = datasets.FashionMNIST(
    root=str(data_dir), train=False, download=False, transform=transform,
)
len(train_dataset), len(test_dataset)
```

## Step 4: Inspect One Example

Check one image and one label. The image shape should be `[1, 28, 28]` after the transform.

```python
# TODO: inspect one example
# image, label = train_dataset[0]
# image.shape, image.dtype, label, image.min().item(), image.max().item()

image, label = train_dataset[0]
image.shape, image.dtype, label, image.min().item(), image.max().item()
```

```python
# TODO: create train_loader and test_loader
# train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=0)
# test_loader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=0)
# images, labels = next(iter(train_loader))
# images.shape, labels.shape

train_dataloader = DataLoader(train_dataset, batch_size=8, shuffle=True, num_workers=0)
test_dataloader = DataLoader(test_dataset, batch_size=8, shuffle=False, num_workers=0)

images, labels = next(iter(train_dataloader))
images.shape, labels.shape
```
